```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
import time

print("\n" + "="*70)
print("AUTOMATED LOGIN TEST SUITE - TASK 2")
print("="*70 + "\n")

# Setup Chrome for Colab
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
chrome_options.add_argument('--disable-gpu')

driver = webdriver.Chrome(options=chrome_options)

test_results = []

# ===== TEST 1: VALID LOGIN =====
print("\n" + "-"*70)
print("TEST 1: Valid Login (student / Password123)")
print("-"*70)

try:
    driver.get("https://practicetestautomation.com/practice-test-login/")
    print("✓ Navigated to login page")

    time.sleep(2)

    # Enter username
    username = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.ID, "username"))
    )
    username.clear()
    username.send_keys("student")
    print("✓ Entered username: student")

    # Enter password
    password = driver.find_element(By.ID, "password")
    password.clear()
    password.send_keys("Password123")
    print("✓ Entered password: Password123")

    # Click submit
    submit = driver.find_element(By.ID, "submit")
    submit.click()
    print("✓ Clicked submit button")

    # Wait for success
    time.sleep(3)

    if "logged in successfully" in driver.page_source.lower() or "congratulations" in driver.page_source.lower():
        print("✓✓✓ TEST 1 PASSED ✓✓✓")
        print(f"✓ Current URL: {driver.current_url}")
        test_results.append(("Test 1: Valid Login", "PASSED ✓", "Valid credentials accepted, successful login"))
    else:
        print("✗✗✗ TEST 1 FAILED ✗✗✗")
        print(f"Current URL: {driver.current_url}")
        test_results.append(("Test 1: Valid Login", "FAILED ✗", "Success page not displayed"))

except Exception as e:
    print(f"✗✗✗ TEST 1 ERROR: {str(e)}")
    test_results.append(("Test 1: Valid Login", "ERROR ✗", str(e)))

# ===== TEST 2: INVALID PASSWORD =====
print("\n" + "-"*70)
print("TEST 2: Invalid Password")
print("-"*70)

try:
    driver.get("https://practicetestautomation.com/practice-test-login/")
    print("✓ Navigated to login page")

    time.sleep(2)

    # Enter username
    username = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.ID, "username"))
    )
    username.clear()
    username.send_keys("student")
    print("✓ Entered username: student")

    # Enter WRONG password
    password = driver.find_element(By.ID, "password")
    password.clear()
    password.send_keys("WrongPassword123")
    print("✓ Entered password: WrongPassword123 (intentionally wrong)")

    # Click submit
    submit = driver.find_element(By.ID, "submit")
    submit.click()
    print("✓ Clicked submit button")

    # Wait for error
    time.sleep(3)

    if "invalid" in driver.page_source.lower() or "error" in driver.page_source.lower():
        print("✓✓✓ TEST 2 PASSED ✓✓✓")
        print("✓ Error message displayed as expected")
        test_results.append(("Test 2: Invalid Password", "PASSED ✓", "Invalid credentials rejected with error message"))
    else:
        print("✗✗✗ TEST 2 FAILED ✗✗✗")
        print(f"Current URL: {driver.current_url}")
        test_results.append(("Test 2: Invalid Password", "FAILED ✗", "Error message not displayed"))

except Exception as e:
    print(f"✗✗✗ TEST 2 ERROR: {str(e)}")
    test_results.append(("Test 2: Invalid Password", "ERROR ✗", str(e)))

# ===== TEST 3: EMPTY FIELDS =====
print("\n" + "-"*70)
print("TEST 3: Empty Fields Validation")
print("-"*70)

try:
    driver.get("https://practicetestautomation.com/practice-test-login/")
    print("✓ Navigated to login page")

    time.sleep(2)

    # Click submit WITHOUT entering anything
    submit = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.ID, "submit"))
    )
    submit.click()
    print("✓ Clicked submit with empty fields")

    # Wait for validation
    time.sleep(3)

    if "username" in driver.page_source.lower() or "required" in driver.page_source.lower():
        print("✓✓✓ TEST 3 PASSED ✓✓✓")
        print("✓ Validation message shown")
        test_results.append(("Test 3: Empty Fields", "PASSED ✓", "Empty fields properly validated"))
    else:
        print("✗✗✗ TEST 3 FAILED ✗✗✗")
        test_results.append(("Test 3: Empty Fields", "FAILED ✗", "Validation not working"))

except Exception as e:
    print(f"✗✗✗ TEST 3 ERROR: {str(e)}")
    test_results.append(("Test 3: Empty Fields", "ERROR ✗", str(e)))

# ===== TEST REPORT =====
print("\n\n" + "="*70)
print("TEST EXECUTION REPORT")
print("="*70)

passed = sum(1 for _, status, _ in test_results if "PASSED" in status)
total = len(test_results)

print(f"\nTest Summary:")
print(f"  Total Tests: {total}")
print(f"  Passed: {passed}")
print(f"  Failed: {total - passed}")
print(f"  Success Rate: {(passed/total)*100:.1f}%")

print(f"\nDetailed Results:")
for i, (test_name, status, details) in enumerate(test_results, 1):
    print(f"\n  {i}. {test_name}")
    print(f"     Status: {status}")
    print(f"     Details: {details}")

print("\n" + "="*70 + "\n")

# Close browser
print("Closing browser...")
driver.quit()
print("✓ Browser closed successfully\n")
```
